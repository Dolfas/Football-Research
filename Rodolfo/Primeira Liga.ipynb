```python
#Primeira Liga Dataframe 
primeira_liga_dataframe = uncleaned_data[uncleaned_data['Division'] == 'P1']
primeira_liga_dataframe 
```

#### Selecting the 2025 data

```python
#Basically we iterate over all the values of the dataframe and find all the matches that occured in 2025
#If we wanted more years, say between 2020 and 2025, just change the list in the .isin() function to ['2020','2021','2022','2023','2024']

primeira_liga_2025_dataframe = primeira_liga_dataframe[primeira_liga_dataframe.iloc[:,1].str[:4].isin(['2025'])]
```

```python
primeira_liga_2025_dataframe
```

```python
## Dropping some variables
primeira_liga_2025_dataframe = primeira_liga_2025_dataframe.drop(columns=['Division'])
```

# Extracting Insights 


## Relation between Half-time results and Full-time results

```python
ft_hf_results_primeira_liga_2025 =  primeira_liga_2025_dataframe[['FTResult','HTResult']]
ft_hf_results_primeira_liga_2025
```

```python
# Check if there are any NaN values 
ft_hf_results_primeira_liga_2025.isnull().values.any()
```

```python
ct = pd.crosstab(ft_hf_results_primeira_liga_2025['HTResult'], ft_hf_results_primeira_liga_2025['FTResult'])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7, 7))
im = ax.imshow(ct, cmap='YlOrBr', interpolation='nearest')

ax.set_xticks(range(len(ct.columns)))
ax.set_xticklabels(ct.columns)
ax.set_yticks(range(len(ct.index)))
ax.set_yticklabels(ct.index)
ax.set_xlabel('Full‑time Result')
ax.set_ylabel('Half‑time Result')
fig.colorbar(im, ax=ax, label='Count')

plt.title('HT→FT Result Heat‑map')
plt.show()
```
